```python
import pandas as pd

# 1. CSV 파일 읽기
# 'Untitled.csv' 파일이 코드와 같은 폴더에 있어야 합니다.
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_processing/Untitled.csv');

df
```

```python
# 2. 데이터 준비 및 정제
# '회계년' 컬럼을 숫자 형태로 변환합니다.
df['회계년'] = pd.to_numeric(df['회계년'], errors='coerce')
df.dropna(subset=['회계년'], inplace=True)
df['회계년'] = df['회계년'].astype(int)

# '자본금'과 '매출액'도 숫자로 변환합니다.
df['자본금'] = pd.to_numeric(df['자본금'], errors='coerce')
df['매출액'] = pd.to_numeric(df['매출액'], errors='coerce')
```

```python
# 3. 데이터 필터링
# 2023년과 2024년 데이터만 각각 추출합니다.
df_2023 = df[df['회계년'] == 2023].copy()
df_2024 = df[df['회계년'] == 2024].copy()


# 2024년 기준 자본금 1000억 이상인 기업만 필터링합니다.
df_2024_filtered = df_2024[df_2024['자본금'] >= 100000000]
df_2024_filtered
```

```python
# 4. 2023년과 2024년 데이터 합치기
# '종목코드'를 기준으로 두 데이터를 합쳐서 23년과 24년 매출액을 나란히 놓습니다.
merged_df = pd.merge(
    df_2024_filtered[['종목코드', '종목명', '매출액']],
    df_2023[['종목코드', '매출액']],
    on='종목코드',
    suffixes=('_2024', '_2023')
)
```

```python
# 5. 매출액 증가율 계산
# 2023년 매출액이 0보다 큰 경우에만 계산하여 오류를 방지합니다.
merged_df = merged_df[merged_df['매출액_2023'] > 0]
merged_df['매출액 증가율'] = ((merged_df['매출액_2024'] - merged_df['매출액_2023']) / merged_df['매출액_2023']) * 100
```

```python
# 6. 최종 결과 정렬 및 선택
# '매출액 증가율' 기준으로 내림차순 정렬 후 상위 20개를 선택합니다.
top_20_growth = merged_df.sort_values(by='매출액 증가율', ascending=False).head(20)

# 보기 좋게 컬럼 이름을 변경하고 순서를 정리합니다.
final_output = top_20_growth.rename(columns={
    '매출액_2023': '23년 매출액',
    '매출액_2024': '24년 매출액'
})[['종목코드', '종목명', '23년 매출액', '24년 매출액', '매출액 증가율']]
```

```python
# 7. 결과 출력 및 파일 저장
print("--- 상위 20개 기업 매출액 증가율 ---")
print(final_output)
```
